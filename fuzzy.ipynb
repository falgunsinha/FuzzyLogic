## Task 4

According to this tutorial create fuzzy system.
https://pythonhosted.org/scikit-fuzzy/auto_examples/plot_tipping_problem_newapi.html


Create rules for our fuzzy variable window, temperature, heating and observe the fuzzy output.

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

temp_x = np.linspace(0, 100, 650)

# Fuzzy sets for temperature
very_high = fuzz.trapmf(temp_x, [95, 100, 100, 100])
high = fuzz.trapmf(temp_x, [70, 80, 90, 100])
warm = fuzz.trapmf(temp_x, [12, 25, 60, 80])
low = fuzz.trapmf(temp_x, [0, 5, 10, 15])
very_low = fuzz.trapmf(temp_x, [0, 0, 0, 5])

# Fuzzy sets for window
closed = fuzz.trapmf(temp_x, [0, 0, 20, 40])
partially_open = fuzz.trapmf(temp_x, [20, 40, 60, 80])
open_window = fuzz.trapmf(temp_x, [60, 80, 100, 100])

# Fuzzy sets for heating
very_low_heating = very_low
low_heating = low
medium_heating = warm
high_heating = high
very_high_heating = very_high

# Define fuzzy rules
rule1 = np.fmax(low_heating, np.fmax(closed, very_low))
rule2 = np.fmax(medium_heating, partially_open)
rule3 = np.fmax(high_heating, open_window)
rule4 = very_high_heating

# Aggregate all the rules
aggregated = np.fmax.reduce([rule1, rule2, rule3, rule4])

# Plot the fuzzy sets and the aggregated result
plt.figure(figsize=(10, 6))

plt.plot(temp_x, low_heating, label='Low Heating')
plt.plot(temp_x, medium_heating, label='Medium Heating')
plt.plot(temp_x, high_heating, label='High Heating')
plt.plot(temp_x, very_high_heating, label='Very High Heating')
plt.plot(temp_x, closed, label='Closed Window')
plt.plot(temp_x, partially_open, label='Partially Open Window')
plt.plot(temp_x, open_window, label='Open Window')

plt.plot(temp_x, aggregated, label='Aggregated Output', linestyle='--')

plt.title('Fuzzy Sets and Aggregated Output')
plt.xlabel('Temperature')
plt.ylabel('Membership Value')
plt.legend()

plt.show()
```
